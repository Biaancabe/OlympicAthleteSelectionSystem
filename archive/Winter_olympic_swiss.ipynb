# **1. Introduction**

The selection of Swiss athletes for the Olympic Games is a highly regulated and complex process. In Switzerland, the national selection criteria are defined by Swiss Olympic in collaboration with the relevant sports federations. Each sport applies its own specific requirements, ensuring that only athletes who meet a precise set of conditions are eligible for the Olympic Games.

Currently, these selection criteria are published in PDF format, which makes the verification process time-consuming. When federations propose athletes for the Olympic Games, Swiss Olympic must manually check each candidate against the relevant criteria. This manual approach creates inefficiencies in the process.

For this reason, we were assigned the task of designing a system that translates the selection criteria for the Olympic Winter Games 2026 in Milano Cortina into a machine-readable format (YAML file) and developing a program that uses a list of results in CSV format to automatically check which Swiss athletes meet the selection criteria.

The project will therefore consist of creating a program capable of importing a results list in CSV format and automatically checking which Swiss athletes meet, partially meet, or fail to meet the selection criteria (YAML file). Building on this, a notebook-based interface and a Streamlit interface, that provide an interactive dashboard which allows users to filter by sport and discipline and display detailed information on results.

# **5. Logic for Eligibility**

While implementing the eligibility logic, we encountered differences in the YAML rule schemas across sports. For example, Biathlon uses an additional nesting level called "all_off" that Alpine Skiing does not. To handle this, we needed to add criteria to our eligibility logic.


## **Sport vs. Discipline Filtering**

The code works reliably when filtering by Sport. Following stakeholder feedback (Lionel), discipline-level filtering was also required. To address this, we introduced a Combined_Sport column (Sport + "_" + Discipline), which enables queries at the discipline level and produces lists of eligible athletes for a specific discipline.

At present, discipline-level results are not always consistent. In some cases, there is insufficient data to determine eligibility, while in others certain filters in the code exclude results. These cases have been documented as current limitations.

# **Function Overview**
Below is a concise description of each function.

## **`normalize_filtered_df(df)`**

Cleans and standardizes the data to make later comparisons reliable. It copies the table, converts “Date” to a timezone-free datetime column (Date_dt), coerces “Year” into nullable integers (Year_int), parses “Rank” as numeric (Rank_num), trims whitespace in key text fields (competition, sport, discipline), ensures a “Person” column exists, and constructs a numeric result column (Result_num) from common score fields when available.

## **`get_selection_paths(rules, group="Group_A")`**
Normalizes differently shaped YAML rule files to a single structure: {"any_of": [...]}. If rules contain selection_paths, that block is returned. If rules are grouped under qualification_groups, the requested group (or the group with the lowest priority value) is selected and its selection_paths returned. If rules expose a top-level any_of, it is wrapped accordingly. This removes schema differences for downstream logic.

## **`mask_for_condition(df, cond)`**

Builds a boolean mask marking rows that satisfy one condition. It supports exact competition name matches, specific years, inclusive date intervals, inclusive rank intervals, case-insensitive equality for “Discipline” and “Gender”, and numeric result thresholds via ge, le, or an interval. The mask is later used to isolate rows that meet the specified criteria.

## **`eligible_for_condition(df, mask, count_at_least, entity_col="Person")`**
Determines which entities (athletes) meet a minimum frequency within the masked rows. It counts appearances of entity_col among matches and returns the set whose counts are at least count_at_least.

## **`eval_all_of(df, path, entity_col="Person")`**

Evaluates routes that require multiple conditions simultaneously. A single condition is treated as a one-step list; an all_of list is processed condition by condition. For each step, the eligible set is computed, and the intersection across steps is taken so that only entities satisfying every condition remain.

## **`eval_any_of(df, rules, entity_col="Person")`**
Implements the top-level OR logic. Each entry under any_of—either a single condition or an all_of block—is evaluated, and the union of all resulting eligible sets is returned. An entity qualifies if it meets at least one route.

## **`eligibility_checker(rules, df, sport_key, filter_col="Sport", group="Group_A")`**

Coordinates the workflow. It first filters rows by the chosen sport key in the specified column, then normalizes the filtered data. Next, it selects the appropriate rule block via get_selection_paths (group-aware) and evaluates those rules with eval_any_of. The final output is a sorted list of eligible athlete names.

# **Workflow Summary**
A call provides rules, results table, and an optional sport/discipline filter. The checker narrows the dataset to the targeted sport, standardizes columns for consistent comparisons, normalizes rule schemas to a uniform any_of/all_of structure, and applies the combined AND/OR logic (date, rank, discipline, gender, score thresholds). The output is the list of athletes who satisfy at least one qualifying route, returned as an ordered list.

````python
### Codeblock 7


def normalize_filtered_df(df: pd.DataFrame) -> pd.DataFrame:
    out = df.copy()

    # Dates to datetime
    if "Date" in out.columns:
        out["Date_dt"] = pd.to_datetime(out["Date"], errors="coerce", utc=True).dt.tz_localize(None)
    else:
        out["Date_dt"] = pd.NaT

    # Year to int (coerce invalid to NaN then drop in comparisons)
    if "Year" in out.columns:
        out["Year_int"] = pd.to_numeric(out["Year"], errors="coerce").astype("Int64")
    else:
        out["Year_int"] = pd.Series([pd.NA] * len(out), dtype="Int64")

    # Rank to numeric (strip non-digits like "DNS", "DNF", etc.)
    if "Rank" in out.columns:
        out["Rank_num"] = pd.to_numeric(out["Rank"], errors="coerce")
    else:
        out["Rank_num"] = np.nan

    # Normalize strings we match exactly
    for col in ["Comp.SetDetail", "CompetitionSet", "Sport", "Discipline", "Combined_Sport"]:
        if col in out.columns:
            out[col] = out[col].astype(str).str.strip()

    # Person column (entity key)
    if "Person" not in out.columns:
        out["Person"] = ""

    # create a numeric result column (try common score fields)
    out["Result_num"] = pd.to_numeric(out.get("Result", np.nan), errors="coerce")
    if out["Result_num"].isna().all():
        for alt in ["Score", "Points", "Total", "TotalScore", "Total_Score"]:
            if alt in out.columns:
                out["Result_num"] = out["Result_num"].fillna(pd.to_numeric(out[alt], errors="coerce"))

    return out

# make sure all Sports work, even if the yaml files have different formats
def get_selection_paths(rules: dict, group: str | None = "Group_A") -> dict:
    """
    Always return {"any_of": [...]} no matter how the YAML is structured.
    Supports:
      - {"selection_paths": {"any_of": [...]}}
      - {"qualification_groups": {"Group_A": {"priority": 1, "selection_paths": {...}}, ...}}
      - {"any_of": [...]}
    """
    # 1) Direct selection_paths
    if "selection_paths" in rules:
        sp = rules["selection_paths"]
        return sp if isinstance(sp, dict) and "any_of" in sp else {"any_of": sp}

    # 2) Grouped rules
    if "qualification_groups" in rules:
        groups = rules["qualification_groups"] or {}
        # explicit group if provided
        if group and group in groups:
            sp = groups[group].get("selection_paths", {})
            return sp if "any_of" in sp else {"any_of": sp}
        # else pick lowest priority
        if groups:
            best_name = min(groups, key=lambda name: groups[name].get("priority", float("inf")))
            sp = groups[best_name].get("selection_paths", {})
            return sp if "any_of" in sp else {"any_of": sp}

    # 3) Raw any_of at top-level
    if "any_of" in rules:
        return {"any_of": rules["any_of"]}

    return {"any_of": []}



def mask_for_condition(df: pd.DataFrame, cond: dict) -> pd.Series:
    """
    cond looks like:
      {
        "Comp.SetDetail": "BMW IBU World Cup",
        "Year": 2025,
        "Date": {"interval": ["2025-11-01", "2026-01-18"]},
        "Rank": {"interval": [1, 30]},
        "count_at_least": 1
      }
    """
    #initialize all as True
    mask = pd.Series(True, index=df.index)

    # Check if Competition "BMW IBU World Cup" matches
    if "Comp.SetDetail" in cond:
        mask &= (df.get("Comp.SetDetail", "") == str(cond["Comp.SetDetail"]).strip())

    #Check if the year of the competition matches
    if "Year" in cond:
        # Compare to normalized Int64 year
        target_year = cond["Year"]
        mask &= (df["Year_int"] == target_year)

    # Interval checks (inclusive)
    if "Date" in cond and isinstance(cond["Date"], dict) and "interval" in cond["Date"]:
        start_str, end_str = cond["Date"]["interval"]
        start = pd.to_datetime(start_str, utc=True).tz_localize(None)
        end = pd.to_datetime(end_str, utc=True).tz_localize(None)
        mask &= (df["Date_dt"] >= start) & (df["Date_dt"] <= end)

    if "Rank" in cond and isinstance(cond["Rank"], dict) and "interval" in cond["Rank"]:
        lo, hi = cond["Rank"]["interval"]
        mask &= (df["Rank_num"] >= lo) & (df["Rank_num"] <= hi)

    # case-insensitive equality for Discipline / Gender, added this part to make Figure Skating work
    if "Discipline" in cond:
        left = df.get("Discipline", "").astype(str).str.strip().str.casefold()
        right = str(cond["Discipline"]).strip().casefold()
        mask &= (left == right)

    if "Gender" in cond:
        left = df.get("Gender", "").astype(str).str.strip().str.casefold()
        right = str(cond["Gender"]).strip().casefold()
        mask &= (left == right)

    # NEW: numeric Result thresholds
    if "Result" in cond and isinstance(cond["Result"], dict):
        r = cond["Result"]
        val = pd.to_numeric(df.get("Result_num", np.nan), errors="coerce")
        if "interval" in r:
            lo, hi = map(float, r["interval"])
            mask &= (val >= lo) & (val <= hi)
        if "ge" in r:
            mask &= (val >= float(r["ge"]))
        if "le" in r:
            mask &= (val <= float(r["le"]))

    return mask



def eligible_for_condition(df: pd.DataFrame, mask: pd.Series, count_at_least: int, entity_col: str = "Person") -> set:
    """
    Given a mask, return set of entities that appear at least count_at_least times.
    """
    hits = df.loc[mask, entity_col].dropna()
    if hits.empty:
        return set()
    counts = hits.value_counts()
    return set(counts[counts >= count_at_least].index)



def eval_all_of(df: pd.DataFrame, path: dict, entity_col: str = "Person") -> set:
    """
    path can be:
      - {"all_of": [{"condition": {...}}, {"condition": {...}}]}
      - {"condition": {...}}   # treat as single-step all_of
    """
    # Normalize: if a bare "condition" is provided, treat it as all_of with one step
    if "condition" in path and "all_of" not in path:
        path = {"all_of": [ {"condition": path["condition"]} ]}

    groups = []
    for step in path.get("all_of", []):
        cond = step.get("condition", {})
        mask = mask_for_condition(df, cond)
        threshold = int(cond.get("count_at_least", 1))
        ok = eligible_for_condition(df, mask, threshold, entity_col=entity_col)
        groups.append(ok)

    if not groups:
        return set()

    eligible = groups[0]
    for s in groups[1:]:
        eligible &= s
    print(f"eval_all_of: found {len(eligible)} eligible entities across {len(groups)} groups.")
    return eligible


def eval_any_of(df: pd.DataFrame, rules: dict, entity_col: str = "Person") -> set:
    """
    rules: {"any_of": [ {"condition": {...}}, {"all_of": [...]}, ... ]}
    Returns union of entities that satisfy at least one path.
    """
    winners = set()
    for path in rules.get("any_of", []):
        # Delegate to eval_all_of which now handles both shapes
        winners |= eval_all_of(df, path, entity_col=entity_col)
    return winners



def eligibility_checker(
    rules: dict,
    df: pd.DataFrame,
    sport_key: str,
    filter_col: str = "Combined_Sport",          # set to "Combined_Sport" if that's what you're passing
    group: str | None = "Group_A",
) -> list[str]:

    # 1) filter rows for this sport
    filtered = df[df[filter_col].astype(str).str.strip() == str(sport_key).strip()].copy()
    if filtered.empty:
        return []

    # 2) normalize columns
    filtered = normalize_filtered_df(filtered)

    # 3) pick the correct selection_paths (handles top-level any_of, selection_paths, or qualification_groups)
    selection_paths = get_selection_paths(rules, group=group)

    # 4) evaluate rules
    eligible_set = eval_any_of(filtered, selection_paths, entity_col="Person")

    # 5) stable list
    return sorted(eligible_set)

#chose wether you want to filter for Sport or Discipline, then chose specific Sport/Discipline
filter_col = "Combined_Sport"
sport_key = "Alpine Skiing_Downhill"
rules = Alpine_Skiing_rules
group = "Group_A"
eligible_athletes = eligibility_checker(rules, df, sport_key, filter_col, group)
print(eligible_athletes)

````

#**6. Eligibility Routes**



##**6.1 Enhanced Qualification Path Function**


The get_athletes_by_qualification_path in codeblock 8 is robust because it accepts the full rules dict and uses get_selection_paths(...) to normalize all common YAML shapes—including plain any_of, nested selection_paths, and grouped qualification_groups like Group_A. It also lets the caller choose the filter column (e.g., "Sport" or "Combined_Sport") instead of hard-coding one, produces readable path labels by joining condition descriptions with “ + ”, and reuses the same shared helpers (normalize_filtered_df, get_selection_paths, eval_all_of) as the main eligibility flow for consistent behavior. In practice, this means broader schema compatibility, clearer output, and more informative errors across sports whose YAMLs aren’t identical.

````python
### Codeblock 8
#Adjusted get_athletes_by_qualification_path
def get_athletes_by_qualification_path(
    rules: dict,
    df: pd.DataFrame,
    sport_key: str,
    filter_col: str = "Combined_Sport",      # set to "Combined_Sport" if needed
    group: str | None = "Group_A",  # used when rules have qualification_groups
    entity_col: str = "Person",
) -> dict[str, list[str]]:
    """
    Return {path_name: [sorted athletes qualifying via that path]}.
    """

    # 1) filter rows for this sport/discipline
    filtered = df[df[filter_col].astype(str).str.strip() == str(sport_key).strip()].copy()
    if filtered.empty:
        return {"error": f"No data found for '{sport_key}' in column '{filter_col}'."}

    # 2) normalize columns (dates, year, rank, scores, strings)
    filtered = normalize_filtered_df(filtered)

    # 3) normalize rules to {"any_of": [...]} using existing helper
    selection_paths = get_selection_paths(rules, group=group)
    any_of_paths = selection_paths.get("any_of", [])

    # 4) evaluate each path separately (reuse eval_all_of)
    path_results: dict[str, list[str]] = {}

    def _name(path: dict, idx: int) -> str:
        # prefer human-readable description if present
        if "condition" in path:
            d = path["condition"].get("description")
            if isinstance(d, str) and d.strip():
                return d.strip()
        if "all_of" in path:
            parts = [step.get("condition", {}).get("description", "").strip()
                     for step in path.get("all_of", [])]
            parts = [p for p in parts if p]
            if parts:
                return " + ".join(parts)
        return f"Path_{idx}"

    for i, path in enumerate(any_of_paths, start=1):
        path_name = _name(path, i)
        try:
            elig_set = eval_all_of(filtered, path, entity_col=entity_col)
            path_results[path_name] = sorted(elig_set)
        except Exception as e:
            path_results[path_name] = []
            print(f"Error processing {path_name}: {e}")

    return path_results

  #test with biathlon

print("Test Qualification by Path function")

sport_key = "Cross-Country Skiing_Sprint Freestyle"

results = get_athletes_by_qualification_path(
    rules=Cross_Country_Skiing_rules,  # <-- full dict, not ["selection_paths"]
    df=df,
    sport_key=sport_key,
    filter_col="Combined_Sport",
    group="Group_A",
)

if "error" in results:
    print(f" {results['error']}")
else:
    print("SIMPLE PATH RESULTS:")
    print("-" * 30)
    for path_name, athletes in results.items():
        print(f"{path_name}: {athletes}")

    print(f"\n SUMMARY:")
    total_athletes = sum(len(athletes) for athletes in results.values())
    print(f"Total athletes found across all paths: {total_athletes}")

    # Show which paths have athletes
    active_paths = [path for path, athletes in results.items() if athletes]
    print(f"Paths with qualifying athletes: {active_paths}")
````

````python
### Codeblock 10
##!!!!Note: to run this part of the code you need to install Streamlit and required dependencies
!pip install streamlit
!pip install pyngrok
!pip install pandas numpy PyYAML
# 🏔️ Swiss Olympic Team Selection - Simple Interface
import pandas as pd
import numpy as np
import yaml
import os
from typing import Dict, List, Set, Any, Tuple

# Initialize global variables
notebook_df = None
notebook_data_loaded = False

# Data loading function
def load_olympic_data_notebook():
    """Load Olympic data from CSV file"""
    global notebook_df, notebook_data_loaded
    csv_path = "oly_data/Results_Test_Version.csv"

    try:
        if not os.path.exists(csv_path):
            print(f"❌ File not found: {csv_path}")
            return False

        notebook_df = pd.read_csv(csv_path, sep=";", on_bad_lines="skip")
        notebook_df["Combined_Sport"] = notebook_df["Sport"].astype(str) + "_" + notebook_df["Discipline"]
        notebook_data_loaded = True

        print("✅ Data loaded successfully!")
        print(f"📊 Total records: {len(notebook_df)}")
        print(f"🏃‍♂️ Unique athletes: {notebook_df['Person'].nunique()}")
        print(f"🏅 Sports: {len(notebook_df['Sport'].unique())}")

        return True

    except Exception as e:
        print(f"❌ Error loading data: {e}")
        return False

# Eligibility checking functions (from your notebook)
def normalize_filtered_df(df: pd.DataFrame) -> pd.DataFrame:
    """Normalize data types for filtering operations"""
    out = df.copy()

    # Dates to datetime
    if "Date" in out.columns:
        out["Date_dt"] = pd.to_datetime(out["Date"], errors="coerce", utc=True).dt.tz_localize(None)
    else:
        out["Date_dt"] = pd.NaT

    # Year to int
    if "Year" in out.columns:
        out["Year_int"] = pd.to_numeric(out["Year"], errors="coerce").astype("Int64")
    else:
        out["Year_int"] = pd.Series([pd.NA] * len(out), dtype="Int64")

    # Rank to numeric
    if "Rank" in out.columns:
        out["Rank_num"] = pd.to_numeric(out["Rank"], errors="coerce")
    else:
        out["Rank_num"] = np.nan

    # Normalize strings
    for col in ["Comp.SetDetail", "CompetitionSet", "Sport", "Discipline", "Combined_Sport"]:
        if col in out.columns:
            out[col] = out[col].astype(str).str.strip()

    # Person column
    if "Person" not in out.columns:
        out["Person"] = ""

    return out

def mask_for_condition(df: pd.DataFrame, cond: dict) -> pd.Series:
    """Create boolean mask for a single condition"""
    mask = pd.Series(True, index=df.index)

    if "Comp.SetDetail" in cond:
        mask &= (df.get("Comp.SetDetail", "") == str(cond["Comp.SetDetail"]).strip())

    if "Year" in cond:
        target_year = cond["Year"]
        mask &= (df["Year_int"] == target_year)

    if "Date" in cond and isinstance(cond["Date"], dict) and "interval" in cond["Date"]:
        start_str, end_str = cond["Date"]["interval"]
        start = pd.to_datetime(start_str, utc=True).tz_localize(None)
        end = pd.to_datetime(end_str, utc=True).tz_localize(None)
        mask &= (df["Date_dt"] >= start) & (df["Date_dt"] <= end)

    if "Rank" in cond and isinstance(cond["Rank"], dict) and "interval" in cond["Rank"]:
        lo, hi = cond["Rank"]["interval"]
        mask &= (df["Rank_num"] >= lo) & (df["Rank_num"] <= hi)

    return mask

def eligible_for_condition(df: pd.DataFrame, mask: pd.Series, count_at_least: int, entity_col: str = "Person") -> set:
    """Return entities that meet the count threshold"""
    hits = df.loc[mask, entity_col].dropna()
    if hits.empty:
        return set()
    counts = hits.value_counts()
    return set(counts[counts >= count_at_least].index)

def eval_all_of(df: pd.DataFrame, path: dict, entity_col: str = "Person") -> set:
    """Evaluate "all_of" conditions (intersection)"""
    groups = []
    for step in path.get("all_of", []):
        cond = step.get("condition", {})
        mask = mask_for_condition(df, cond)
        threshold = int(cond.get("count_at_least", 1))
        ok = eligible_for_condition(df, mask, threshold, entity_col=entity_col)
        groups.append(ok)

    if not groups:
        return set()

    eligible = groups[0]
    for s in groups[1:]:
        eligible &= s
    return eligible

def eval_any_of(df: pd.DataFrame, rules: dict, entity_col: str = "Person") -> set:
    """Evaluate "any_of" conditions (union)"""
    winners = set()
    for path in rules.get("any_of", []):
        winners |= eval_all_of(df, path, entity_col=entity_col)
    return winners

def check_eligibility_notebook(df: pd.DataFrame, rules: dict, sport_discipline: str) -> Tuple[List, Dict]:
    """Main eligibility checking function"""
    try:
        # Filter data for the specific sport/discipline
        filtered = df[df["Combined_Sport"].astype(str).str.strip() == str(sport_discipline).strip()].copy()

        if filtered.empty:
            return [], {"error": f"No data found for {sport_discipline}"}

        # Normalize data
        filtered = normalize_filtered_df(filtered)

        # Evaluate rules
        selection_paths = rules.get("selection_paths", {})
        eligible_set = eval_any_of(filtered, selection_paths, entity_col="Person")

        # Prepare detailed results
        results = {
            "total_records": len(filtered),
            "eligible_count": len(eligible_set),
            "sport_discipline": sport_discipline,
            "athletes_data": filtered[filtered["Person"].isin(eligible_set)] if eligible_set else pd.DataFrame()
        }

        return sorted(eligible_set), results

    except Exception as e:
        return [], {"error": str(e)}

def load_yaml_rules_notebook(sport_name: str) -> Dict:
    """Load YAML rules for a specific sport"""
    yaml_files = {
        "Biathlon": "Biathlon_Hauptkriterien.txt",
        "Alpine Skiing": "Alpine_Skiing_Hauptkriterien.txt",
        "Freestyle Skiing": "Freestyle_Skiing_Hauptkriterien.txt",
        "Figure Skating": "Figure_Skating_Hauptkriterien.txt",
        "Cross Country Skiing": "Cross_Country_Skiing_Hauptkriterien.txt",
        "Bobsleigh": "Bobsleigh_Hauptkriterien.txt"
    }

    if sport_name not in yaml_files:
        return {}

    yaml_path = os.path.join("oly_data/", yaml_files[sport_name])

    try:
        if not os.path.exists(yaml_path):
            return {}

        with open(yaml_path, "r", encoding="utf-8-sig") as file:
            text = file.read().replace("\t", "  ")
            rules = yaml.safe_load(text)
            return rules

    except Exception as e:
        print(f"Error loading YAML for {sport_name}: {e}")
        return {}

# Simple Interface Functions (No Widgets)
def show_athlete_details(athlete_name):
    """Show detailed athlete information"""
    if not notebook_data_loaded:
        print("❌ Please load data first using load_olympic_data_notebook()")
        return

    athlete_data = notebook_df[notebook_df["Person"] == athlete_name]

    print(f"🏃‍♂️ ATHLETE DETAILS: {athlete_name}")
    print("=" * 50)
    print(f"📊 Total Competitions: {len(athlete_data)}")
    print(f"🏅 Sports: {len(athlete_data['Sport'].unique())}")
    print(f"🌍 Countries: {len(athlete_data['Country'].unique())}")

    # Best rank
    ranks = pd.to_numeric(athlete_data["Rank"], errors="coerce").dropna()
    if not ranks.empty:
        print(f"🥇 Best Rank: {ranks.min()}")
        print(f"📉 Worst Rank: {ranks.max()}")

    print("\n🏅 SPORTS:")
    for sport in athlete_data["Sport"].unique():
        print(f"  • {sport}")

    print("\n🎯 DISCIPLINES:")
    for discipline in athlete_data["Discipline"].unique():
        print(f"  • {discipline}")

    print("\n📅 RECENT COMPETITIONS:")
    recent_data = athlete_data[["Date", "Competition", "Sport", "Discipline", "Rank", "Country"]].head(10)
    display(recent_data)

def check_eligibility_simple(sport, discipline):
    """Check athlete eligibility for a sport/discipline combination"""
    if not notebook_data_loaded:
        print("❌ Please load data first using load_olympic_data_notebook()")
        return

    print(f"🔄 Checking eligibility for {sport} - {discipline}...")

    # Load YAML rules
    rules = load_yaml_rules_notebook(sport)

    if rules:
        # Run eligibility check
        combined_sport = f"{sport}_{discipline}"
        eligible_athletes, results = check_eligibility_notebook(notebook_df, rules, combined_sport)

        # Display results
        if "error" not in results:
            print(f"🏆 ELIGIBILITY RESULTS FOR {sport.upper()} - {discipline.upper()}")
            print("=" * 60)
            print(f"📊 Records Analyzed: {results['total_records']}")
            print(f"✅ Eligible Athletes: {results['eligible_count']}")
            print()

            if eligible_athletes:
                print("🥇 ELIGIBLE ATHLETES:")
                print("-" * 30)
                for i, athlete in enumerate(eligible_athletes, 1):
                    print(f"{i}. {athlete}")

                # Show detailed data
                if not results["athletes_data"].empty:
                    print("\n📊 COMPETITION DETAILS:")
                    print("-" * 30)
                    display_cols = ["Person", "Date", "Competition", "Rank", "Result"]
                    available_cols = [col for col in display_cols if col in results["athletes_data"].columns]
                    display(results["athletes_data"][available_cols])
            else:
                print("❌ No eligible athletes found for the current criteria.")
        else:
            print(f"❌ {results['error']}")
    else:
        print(f"❌ Could not load qualification rules for {sport}")

def list_athletes():
    """List all available athletes"""
    if not notebook_data_loaded:
        print("❌ Please load data first using load_olympic_data_notebook()")
        return

    athletes = sorted(notebook_df["Person"].dropna().unique())
    print(f"🏃‍♂️ Available Athletes ({len(athletes)} total):")
    print("=" * 40)
    for i, athlete in enumerate(athletes, 1):
        print(f"{i:3d}. {athlete}")

def list_sports():
    """List all available sports and their disciplines"""
    if not notebook_data_loaded:
        print("❌ Please load data first using load_olympic_data_notebook()")
        return

    sports = sorted(notebook_df["Sport"].unique())
    print(f"🏅 Available Sports ({len(sports)} total):")
    print("=" * 40)

    for sport in sports:
        disciplines = sorted(notebook_df[notebook_df["Sport"] == sport]["Discipline"].unique())
        print(f"\n🏔️ {sport}:")
        for discipline in disciplines:
            print(f"   • {discipline}")

# Main interface function
def show_interface():
    """Show the main interface instructions"""
    print("🏔️ Swiss Olympic Team Selection Interface")
    print("=" * 50)
    print()
    print("📋 Available Functions:")
    print("1. load_olympic_data_notebook() - Load the Olympic data")
    print("2. list_athletes() - Show all available athletes")
    print("3. show_athlete_details('Athlete Name') - Show athlete details")
    print("4. list_sports() - Show all sports and disciplines")
    print("5. check_eligibility_simple('Sport', 'Discipline') - Check eligibility")
    print()
    print("🎯 Example Usage:")
    print("   load_olympic_data_notebook()")
    print("   list_athletes()")
    print("   show_athlete_details('Lena Häcki-Groß')")
    print("   check_eligibility_simple('Biathlon', '12.5km Mass Start')")
    print()
    print("✅ Interface ready!")

# Show the interface
show_interface()
````

#**9. Conclusion**

This project demonstrates that Swiss Olympic’s selection criteria, originally published in PDFs, can be converted into a structured machine-readable format (YAML) and systematically applied to real competition data (CSV) to assess athlete eligibility.

The algorithm developed is capable of handling consistent schema designs, including grouped conditions and mixed AND/OR logic. It standardizes event data (dates, ranks, results) and evaluates athletes against the rules to classify them as eligible, partially eligible, or not eligible.

To extend functionality, discipline-level filtering was implemented by combining sport and discipline into a single key, along with transparent qualification paths that clearly indicate how an athlete qualified.

The implementation of the interactive interfaces supports quick lookups by sport, discipline, and athlete. This solution reduces manual review effort and ensures consistency and traceability in selection decisions for the Milano–Cortina 2026 Olympics. While current limitations include incomplete discipline-level data and varying schema complexity, the approach remains scalable and ready for future extension.


#**10. Challenges and Reflection**

One of the main challenges in this assignment was the coordination and assignment of the workload between the four group members. We were quite fast creating a quick mvp with Generative AI consisting of a dashboard in streamlit where the user could pick a sport and see the results of all athletes qualifying for the sport. However the different disciplines within the sport were not considered. From there we splitted the work into the following parts.

1. Loading the data and transform yaml files into a dictionary
1. Datapreparation
2. Creation of YAML files for all sports
3. Filtering Logic
4. Streamlit Dashboard
5. General documentation / Introduction

Handling differences in the availability and structure of eligibility data was another Challenge we faced. While four out of six sports used a consistent YAML format, making their integration relatively straightforward, the remaining two required significantly more effort due to entirely new structure levels. To address this, we extended the base eligibility logic to incorporate the grouping requirements of Freestyle Skiing. For Figure Skating, which relies on score-based criteria, we chose a more pragmatic approach by adapting its YAML file to align with the existing templates. This solution allowed us to unify the logic across all sports and generate accurate results. Ultimately, the project highlights the importance of flexible data handling and pragmatic design choices when integrating diverse eligibility rules.

Since our project was built in a modular way, we had to make sure the different parts (data processing, rule engine, and UI) worked well together. To keep things simple, we defined clear connections between the components early on, which let us work on them in parallel without constantly checking in with each other. We also ran regular integration tests, which helped us spot small issues early before they could turn into bigger problems later.

A final challenge was ensuring a lean and accessible way to present the results to stakeholders. Our goal was to minimize the need for them to interact directly with code while still allowing transparency and reproducibility. In the end, we settled on using Google Colab, which hosts all the code and integrates both a Streamlit app and an output-based interface. With this setup, stakeholders can simply be invited to the Colab, inspect the underlying implementation if they wish, and run the Streamlit app—hosted via Cloudflare’s free tier—without any additional setup.

# **Limitations**

However, the project also comes with some limitations. Several disciplines currently do not yield results because of insufficient data points, which restricts the eligibility logic. Since the application filters at the discipline level, athletes must meet all conditions within a single discipline, which further limits the applicability.

Another limitation is that adding new athlete data or extending the application to additional sports or disciplines requires some coding knowledge. Ideally, such functionality should be integrated directly into the interface so that non-technical users can upload data or expand coverage without modifying the code. Due to limited time, these features were not implemented in the current version.

Lastly, in a production-ready version, the app should present richer athlete profiles—showing each athlete’s discipline(s), the specific races entered, event dates and competitions, and detailed outcomes (rank, score, etc.). This would reveal not only who qualified but also the exact performances behind every qualification.
